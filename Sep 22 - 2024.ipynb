```python
def create_data_1(m, n=1):
    np.random.seed(10)
    X = np.random.random((m, n))
    w = np.random.randint(-20, 20, (1, n+1))
    ones = np.ones((m,1))
    X1 = np.c_[X, ones]
    print("Generated weights: ", w)
    y = X1.dot(w.T) + 2*np.random.random((m, 1)) - 2*np.random.random((m, 1))
    return X, y
```

```python
X, y = create_data_1(1000, 3)
```

```python
def loss(X, y, w):
    yp = X.dot(w.T)
    diff = yp - y
    sqdiff = diff**2
    return np.sum(sqdiff, axis=0)
```

```python
def gradients(X,  y, w):
    initialloss = loss(X, y, w)[0]
    print("Initial Loss: ", initialloss)
    delta = 1e-10
    lw = w.shape[1]
    new_ws = np.diag(np.zeros(lw) + delta) + w
    new_losses = loss(X, y, new_ws)
    print("new_losses: ", new_losses)
    return (new_losses - initialloss)/delta
```

```python
eta = 0.0001

# X1 = add 1 to X.

w = np.random.random((1, n+1))
```

```python
for epoch in range(1000):
    print("===== EPOCH =======", epoch)
    #compute gradients
    grads = gradients(X, y , w)
    print("grads: ", grads)

    # update the weights in proporation to the gradient
    w = w - eta * grads
    print("Final Loss: ", loss(X, y, w))
```
